### Ejercicio 2

##### Muestra 5 lanzamientos cuya plataforma empiece por 'CCS'


```python
query = """select LaunchSite from SPACEXTBL where LaunchSite like '%CCS%' order by LaunchSite DESC limit 5 ;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 3

##### Mustra el la masa total transportada por los lanzamientos realizados donde el cliente es NASA (CRS)


```python
query = """select sum(PayLoadMass) as TotalMasa from SPACEXTBL where Customers like '%NASA (CRS)%'  ;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 4

##### Muestra la masa media tranportada por lanzamientos con serie que comienza por B1

```python
query = """select LaunchSite,Serial,avg(PayLoadMass) as PesoMediopoSerial from SPACEXTBL where Serial like 'B1%' group by LaunchSite  ;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 5

##### Muestra la fecha donde se realizo el primer aterrizaje exitoso sobre terreno (TRUE RTLS)

*Pista:Utiliza la función min*


```python
query = """ select Date,LandingOutcome from SPACEXTBL where LandingOutcome like '%True RTLS%' order by Date ASC limit 1;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 6

##### Muestra las series de los cohetes que aterrizaron sobre portaviones (TRUE ASDS) con una masa entre 4000 y 6000

```python
query = """ select Serial,PayloadMass from SPACEXTBL where (LandingOutcome like '%True ASDS%') and (PayloadMass	BETWEEN 4000 and 6000) ;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 7

##### Muestra el numero de lanzamientos que completaron su mision correctamente y aquellos que fallaron

```python
query = """ select CASE WHEN Class == 1 THEN Class END AS intentos_buenos,
            CASE WHEN Class == 0 THEN Class END AS intentos_fallidos,
            COUNT(intentos_buenos), COUNT (intentos_fallidos)
            FROM SPACEXTBL
            GROUP BY Class 
            HAVING NOT Class NULL ;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 8

##### Muestra las series de los cohetes que transportaron la masa máxima

```python
query = """ select Serial, max(PayloadMass) from SPACEXTBL GROUP BY Serial ;"""
df  = pd.read_sql(query, con = engine )
df
```
